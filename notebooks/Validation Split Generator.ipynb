```python
from bing_image_downloader import downloader
```

```python
def download_images(query_str, limit, output_dir='../validation/'):
    downloader.download(
        query_str, 
        limit=limit, 
        output_dir=output_dir,
        adult_filter_off=True,
        force_replace=False,
        timeout=60
    )
```

```python
classes = [
    'angry human face', 
    'contempt human face', 
    'disgust human face', 
    'fear human face', 
    'happy human face', 
    'neutral human face', 
    'sad human face', 
    'surprise human face'
]

num_images_per_class = 50
```

```python
for class_name in classes:
    download_images(class_name, num_images_per_class)
```
